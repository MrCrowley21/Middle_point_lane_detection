```python
def apply_mask(image):
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    lower_white = np.array([0, 0, 200])
    higher_white = np.array([255, 255, 255])
    mask = cv2.inRange(hsv_image, lower_white, higher_white)
    masked_image = cv2.bitwise_and(image, image, mask=mask)
    return masked_image
```

```python
def blur_gray_image(color_isolated):
    gray_image = cv2.cvtColor(color_isolated, cv2.COLOR_RGB2GRAY)
    kernel_size = 7
    kernel = np.ones((7, 7), np.uint8)
    blur_gray = cv2.GaussianBlur(gray_image, (kernel_size, kernel_size), 0)
    blur_gray = cv2.dilate(blur_gray, kernel, iterations=1)
    return blur_gray
```

```python
def detect_edges(image):
    low_treshold = 50
    high_treshold = 150
    image_edges = cv2.Canny(image, low_treshold, high_treshold)
    return image_edges
```

```python
def limit_region(edges, image):
    mask = np.zeros_like(edges)
    ignore_mask_color = 255
    imshape = image.shape
    vertices = np.array([[
        (0, imshape[0]),
        (imshape[1] * 0.25, imshape[0] * 0.5),
        (imshape[1] * 0.75, imshape[0] * 0.5),
        (imshape[1], imshape[0])]], dtype=np.int32)
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    masked_edges = cv2.bitwise_and(edges, mask)
    return masked_edges
```

```python
def get_lane_equations(masked_image):
    mxb = np.array([[0, 0]])
    rho = 2
    theta = np.pi / 180
    treshold = 40
    min_line_len = 20
    max_line_gap = 1
    lines = cv2.HoughLinesP(masked_image, rho, theta, treshold, np.array([]), min_line_len, max_line_gap)
    for line in lines:
        for x1, x2, y1, y2 in line:
            m = (y2 - y1) / (x2 - x1)
            b = y1 + (-1) * m * x1
            mxb = np.vstack((mxb, [m, b]))
    median_right_m = np.median(mxb[mxb[:, 0] > 0, 0])
    median_left_m = np.median(mxb[mxb[:, 0] < 0, 0])
    median_right_b = np.median(mxb[mxb[:, 0] > 0, 1])
    median_left_b = np.median(mxb[mxb[:, 0] < 0, 1])
    return median_right_m, median_left_m, median_right_b, median_left_b
```

```python
def get_middle_point(median_right_m, median_left_m, median_right_b, median_left_b, image):
    x_middle = (median_left_b - median_right_b) / (median_right_m - median_left_m)
    y_middle = median_right_m * (median_left_b - median_right_b) / (median_right_m - median_left_m) + \
    median_right_b
    imshape = image.shape
    left_bottom = (imshape[0] - median_left_b) / median_left_m
    right_bottom = (imshape[0] - median_right_b) / median_right_m
    return x_middle, y_middle, left_bottom, right_bottom
```

```python
masked_image = apply_mask(image)
plt.imshow(masked_image)
```

```python
blur_image = blur_gray_image(masked_image)
plt.imshow(blur_image)
```

```python
image_edges = detect_edges(blur_image)
plt.imshow(image_edges)
```

```python
limited_edges = limit_region(image_edges, image)
plt.imshow(limited_edges)
```

```python
median_right_m, median_left_m, median_right_b, median_left_b = get_lane_equations(limited_edges)
```

```python
x_middle, y_middle, left_bottom, right_bottom = get_middle_point(median_right_m, 
                                                       median_left_m, median_right_b, median_left_b, image)
x_middle
```

```python
point_image = np.copy(image) * 0
imshape = image.shape
# cv2.circle(point_image, (np.int_(y_middle), np.int_(left_bottom)), radius=10, color=(255, 0, 0), thickness=-1)
cv2.line(point_image, (np.int_(left_bottom), imshape[0]), (np.int_(x_middle), np.int_(y_middle)), 
                                                           (255, 0, 0), 10)
cv2.line(point_image, (np.int_(right_bottom), imshape[0]), (np.int_(x_middle), np.int_(y_middle)), 
                                                           (255, 0, 0), 10)
lane_edges = cv2.addWeighted(image, 0.8, point_image, 1, 0)
```
